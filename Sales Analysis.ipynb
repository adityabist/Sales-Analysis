```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
    ###Task 1- Merging 12 months of data into a single file

df=pd.read_csv("Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data/Sales_April_2019.csv")

files=[file for file in os.listdir('Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data')]

all_months_data=pd.DataFrame()

for file in files: 
        df= pd.read_csv("Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data/"+ file)
        all_months_data = pd.concat([all_months_data, df])
all_months_data.to_csv("all_data.csv",index=False)
```

```python
##Clean up the data
#Find and Drop NaN rows

nan_df=all_data[all_data.isna().any(axis=1)]
nan_df.head(8)
all_data=all_data.dropna(how='all')
all_data.isna().sum()
all_data
```

```python
#Find and Drop Or rows
```

```python
tempdata=all_data[all_data['Order Date'].str[0:2]=="Or"]


```

```python
clean_data=all_data[all_data['Order Date'].str[0:2]!="Or"]
clean_data.head()
```

```python
##Task2 Adding a month column
```

```python
clean_data['Month']= clean_data['Order Date'].str[0:2]
clean_data['Month']= clean_data['Month'].astype(int)
clean_data.head()
```

```python
#Task 3 Add a sales column
# Converting QO and PE columns to numerics
clean_data["Price Each"]=pd.to_numeric(clean_data["Price Each"])
clean_data["Quantity Ordered"]=pd.to_numeric(clean_data["Quantity Ordered"])

clean_data["Sales"]=clean_data["Quantity Ordered"]*clean_data["Price Each"]
clean_data.head()
```

```python
clean_data.groupby("Month").sum()
```

```python
## What city did highest sales
##Split cells to get city
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split (" ")[1]
clean_data["City"]=clean_data["Purchase Address"].apply(lambda x:get_city(x)+' ('+get_state(x)+ ')')
clean_data.head()
```

```python
results=clean_data.groupby("City").sum()
```

```python
cities= [city for city,df in all_data.groupby('City')]
##cities= all_data['City'].unique()
plt.bar (cities,results["Sales"])
plt.xticks(cities, rotation="vertical", size=8)
plt.ylabel("Sales in USD ($)")
plt.xlabel ("City names")
plt.show()
```

```python
What products are sold often?
```

```python
df=all_data[all_data["Order ID"].duplicated(keep=False)]

df["Grouped"]=df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))

df=df[["Order ID","Grouped"]].drop_duplicates()
```

```python
df.head()
```

```python
from itertools import combinations
from collections import Counter
count=Counter()
## for counting pairs of twos
for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
for key, value in count.most_common(10):    
    print(key,value)
```

What product sold the most?


```python
productgroup=all_data.groupby('Product')
qtordered=productgroup.sum()['Quantity Ordered']
```

```python
products=[prod for prod,df  in productgroup]

plt.plot(products,qtordered)
plt.xticks(products, rotation="vertical", size=8)
plt.grid()
plt.ylabel("Qt ordered")
plt.xlabel ("Product")
```

```python
prices=all_data.groupby('Product').mean()['Price Each']
prices

fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(products,qtordered,color="r")
ax2.plot(products,prices,"b-")

ax1.set_xlabel("product Name")
ax1.set_ylabel("QT ordered", color='r')
ax2.set_ylabel("Price in USD", color='b')

ax1.set_xticklabels(products, rotation="vertical", size=8)
```
